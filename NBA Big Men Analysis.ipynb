```python
#Convert Height to In 

for i in range (len(df)):
    df.iat[i,3] = df.iat[i,3] / 2.54
    round(df.iat[i,3])
```

```python
#Convert Weight to Lbs
for i in range (len(df)):
    df.iat[i,4] = df.iat[i,4] * 2.205
    round(df.iat[i,4])
df
```

```python
#Round Weight
for i in range (len(df)):
    df.iat[i,4] = round(df.iat[i,4])
```
